```python
import pandas as pd

# Load the data
file_path = r"C:\Users\neba\Desktop\Insurance-Analytics\Data\cleaned_insurance_data.csv"
insurance_data = pd.read_csv(file_path)

# Display the first few rows of the dataset
insurance_data.head()
```

Risk Differences Across Provinces
Null Hypothesis (Hâ‚€): There are no risk differences across provinces.
Alternative Hypothesis (Hâ‚): There are significant risk differences across provinces.
KPI: TotalClaims or TotalPremium
Test: ANOVA (if provinces > 2), T-test (if comparing two provinces)

```python
import scipy.stats as stats

# Group data by province and perform ANOVA
province_groups = [group['TotalClaims'].dropna() for _, group in insurance_data.groupby('Province')]

# Perform ANOVA test
f_stat, p_value = stats.f_oneway(*province_groups)

print("ANOVA Test for Risk Differences Across Provinces")
print(f"F-statistic: {f_stat}, p-value: {p_value}")

if p_value < 0.05:
    print("Reject Null Hypothesis: There are significant risk differences across provinces.")
else:
    print("Fail to Reject Null Hypothesis: No significant risk differences across provinces.")
```

Risk Differences Between Zip Codes
Null Hypothesis (Hâ‚€): There are no risk differences between zip codes.
Alternative Hypothesis (Hâ‚): There are significant risk differences between zip codes.
KPI: TotalClaims or TotalPremium
Test: ANOVA (for multiple zip codes), T-test (for two zip codes)

```python
# Group data by PostalCode and perform ANOVA
postal_groups = [group['TotalClaims'].dropna() for _, group in insurance_data.groupby('PostalCode')]

# Perform ANOVA test
f_stat, p_value = stats.f_oneway(*postal_groups)

print("ANOVA Test for Risk Differences Across Postal Codes")
print(f"F-statistic: {f_stat}, p-value: {p_value}")

if p_value < 0.05:
    print("Reject Null Hypothesis: There are significant risk differences between postal codes.")
else:
    print("Fail to Reject Null Hypothesis: No significant risk differences between postal codes.")
```

```python
# Group data by PostalCode and perform ANOVA
postal_profit_groups = [group['ProfitMargin'].dropna() for _, group in insurance_data.groupby('PostalCode')]

# Perform ANOVA test
f_stat, p_value = stats.f_oneway(*postal_profit_groups)

print("ANOVA Test for Profit Margin Differences Across Postal Codes")
print(f"F-statistic: {f_stat}, p-value: {p_value}")

if p_value < 0.05:
    print("Reject Null Hypothesis: There are significant profit margin differences between postal codes.")
else:
    print("Fail to Reject Null Hypothesis: No significant profit margin differences between postal codes.")
```

Risk Differences Between Women and Men
Null Hypothesis (Hâ‚€): There are no significant risk differences between Women and Men.
Alternative Hypothesis (Hâ‚): There are significant risk differences between Women and Men.
KPI: TotalClaims
Test: T-test (as gender has only two groups)

```python
# Split data into male and female groups
male_group = insurance_data[insurance_data['Title'].str.contains('Mr', na=False)]['TotalClaims']
female_group = insurance_data[insurance_data['Title'].str.contains('Mrs|Ms|Miss', na=False)]['TotalClaims']

# Perform T-test
t_stat, p_value = stats.ttest_ind(male_group.dropna(), female_group.dropna())

print("T-test for Risk Differences Between Men and Women")
print(f"T-statistic: {t_stat}, p-value: {p_value}")

if p_value < 0.05:
    print("Reject Null Hypothesis: There are significant risk differences between Women and Men.")
else:
    print("Fail to Reject Null Hypothesis: No significant risk differences between Women and Men.")
```
