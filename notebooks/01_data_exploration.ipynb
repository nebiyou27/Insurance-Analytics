```python
import numpy as np
import pandas as pd
from scipy.stats.mstats import winsorize

def handle_outliers(df):
    """
    Handles outliers in numerical columns using IQR filtering and winsorization.
    - Winsorization for highly skewed features
    - IQR filtering for moderately skewed features
    - Leaves TotalPremium and TotalClaims unchanged
    """
    
    # Features with high skewness -> Winsorization (capping extreme values)
    winsorize_cols = ["TotalPremium", "CalculatedPremiumPerTerm", "CustomValueEstimate", "TotalClaims"]
    
    for col in winsorize_cols:
        if col in df.columns:
            df[col] = winsorize(df[col], limits=[0.01, 0.01])  # Capping extreme 1% values
    
    # IQR method for moderately skewed columns
    iqr_cols = ["Cylinders", "cubiccapacity", "SumInsured", "VehicleAge", "PostalCode", "PolicyID"]
    
    for col in iqr_cols:
        if col in df.columns:
            Q1 = df[col].quantile(0.25)
            Q3 = df[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    
    return df
```

```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

def comprehensive_data_cleaning(data_path, output_path):
    """
    Performs comprehensive data cleaning, outlier handling, and saves the processed data to a CSV file.

    Args:
        data_path (str): Path to the input CSV file.
        output_path (str): Path to save the processed CSV file.
    """
    try:
        # Inspect TrackingDevice column
        with open(data_path, 'r') as f:
            for i, line in enumerate(f):
                if i == 0:
                    header = line.strip().split(',')
                    try:
                        tracking_device_index = header.index('TrackingDevice')
                        break
                    except ValueError:
                        print("TrackingDevice column not found.")
                        exit()

            values = []
            with open(data_path, 'r') as f2:
                next(f2) #skip header.
                for line in f2:
                    row = line.strip().split(',')
                    values.append(row[tracking_device_index])

            print(f"Unique values in TrackingDevice: {set(values)}")

        # Load Data with error handling for TrackingDevice
        def tracking_device_converter(val):
            try:
                return float(val)
            except ValueError:
                return np.nan

        dtype_dict = {'AlarmImmobiliser': str, 'TrackingDevice': object}
        converters = {'TrackingDevice': tracking_device_converter}
        df = pd.read_csv(data_path, dtype=dtype_dict, converters=converters)

    except FileNotFoundError:
        print(f"Error: File not found at {data_path}")
        return

    # Ensure Date Columns Are Parsed Correctly if Available
    if 'TransactionMonth' in df.columns:
        df['TransactionMonth'] = pd.to_datetime(df['TransactionMonth'], errors='coerce')

    # Check for 'RegistrationYear' column instead of 'TransactionYear'
    if 'RegistrationYear' in df.columns:
        df['RegistrationYear'] = pd.to_numeric(df['RegistrationYear'], errors='coerce')

    # Clean non-numeric values and whitespace in numeric columns
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    for col in numeric_cols:
        df[col] = df[col].astype(str).str.strip().replace('', np.nan)
        df[col] = pd.to_numeric(df[col], errors='coerce')

    # Handling Infinite Values
    pd.set_option('future.no_silent_downcasting', True)
    df.replace([np.inf, -np.inf], np.nan, inplace=True)
    pd.set_option('future.no_silent_downcasting', False)

    # Calculate VehicleAge using TransactionMonth year if available, otherwise current year
    if 'TransactionMonth' in df.columns:
        df['VehicleAge'] = df['TransactionMonth'].dt.year - df['RegistrationYear']
    else:
        print("TransactionMonth column not found, defaulting to current year.")
        current_year = pd.to_datetime('today').year
        df['VehicleAge'] = current_year - df['RegistrationYear']

    # Drop columns with excessive missing values (>50%)
    drop_cols = ["WrittenOff", "Rebuilt", "Converted", "CrossBorder", "NumberOfVehiclesInFleet"]
    df.drop(columns=drop_cols, inplace=True, errors='ignore')

    # Fill missing values for categorical columns with mode (avoid chained assignment)
    categorical_cols = df.select_dtypes(include=['object']).columns
    for col in categorical_cols:
        mode_series = df[col].mode()
        if not mode_series.empty:
            df[col] = df[col].fillna(mode_series[0])
        else:
            print(f"Warning: Mode is empty for column {col}. Filling with 'Unknown'.")
            df[col] = df[col].fillna('Unknown')

    # Fill missing values for numerical columns with median (avoid chained assignment)
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    for col in numeric_cols:
        if not df[col].isnull().all(): #check for empty columns.
            df[col] = df[col].fillna(df[col].median())

    # Drop rows where critical financial values are missing
    df.dropna(subset=["CapitalOutstanding"], inplace=True)

    # Convert categorical variables to proper types
    categorical_cols = df.select_dtypes(include=['object']).columns
    df[categorical_cols] = df[categorical_cols].astype("category")

    # Ensure non-negative values for financial columns
    df.loc[df["TotalClaims"] < 0, "TotalClaims"] = 0
    df.loc[df["TotalPremium"] < 0, "TotalPremium"] = 0

    # Handling Outliers
    skewed_cols = ["TotalPremium", "CalculatedPremiumPerTerm", "CustomValueEstimate", "TotalClaims"]
    for col in skewed_cols:
        if (df[col] > 0).all():
            df[col] = np.log1p(df[col])
        else:
            print(f"Column {col} contains non-positive values, log transformation skipped.")

    q1 = df["VehicleAge"].quantile(0.25)
    q3 = df["VehicleAge"].quantile(0.75)
    iqr = q3 - q1
    upper_bound = q3 + 1.5 * iqr
    lower_bound = q1 - 1.5 * iqr

    df["VehicleAge"] = np.where(df["VehicleAge"] > upper_bound, upper_bound, df["VehicleAge"])
    df["VehicleAge"] = np.where(df["VehicleAge"] < lower_bound, lower_bound, df["VehicleAge"])

    boxcox_cols = ['cubiccapacity']
    for col in boxcox_cols:
        if (df[col] > 0).all():
            df[col], _ = stats.boxcox(df[col])
        else:
            print(f"Column {col} contains non-positive values, Box-Cox transformation skipped.")

    # Save the processed DataFrame to a new CSV file
    df.to_csv(output_path, index=False)

    print(f"Processed data saved to: {output_path}")

    # Visualizing CustomValueEstimate after outlier handling
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='CustomValueEstimate', data=df)
    plt.title('Boxplot of CustomValueEstimate After Outlier Handling')
    plt.show()

data_path = r"C:\Users\neba\Documents\week-3\MachineLearningRating_v3.csv"
output_path = r"C:\Users\neba\Documents\week-3\cleaned_insurance_data.csv"
comprehensive_data_cleaning(data_path, output_path)
```
